```python
import pandas as pd

# ダイエットブログ
blogs = pd.read_csv('/home/is/sakiho-k/research/data/diet_data.csv')
# 摂食障害ブログ
disorder = pd.read_csv('/home/is/sakiho-k/research/data/disorder_data.csv')

# テキストがNaNの行を削除
blogs = blogs.dropna(subset=['text'])
disorder = disorder.dropna(subset=['text'])

# "category"列を"labels"列に変更
blogs = blogs.rename(columns={'category': 'labels'})
disorder = disorder.rename(columns={'category': 'labels'})

# "dete"をdatetime型に変換
blogs['date'] = pd.to_datetime(blogs['date'])
disorder['date'] = pd.to_datetime(disorder['date'])

# ユーザーごとにtextを逆順で結合し、sex, age, labels 列を保持
def group_and_concat(df):
    grouped = df.groupby('user').agg({
        'text': lambda x: ' '.join(x[::-1]),
        'sex': 'first',  # 各ユーザーで同じ値を保持
        'age': 'first',  # 各ユーザーで同じ値を保持
        'labels': 'first'  # 各ユーザーで同じ値を保持
    }).reset_index()
    return grouped

# データフレームの処理
blogs = group_and_concat(blogs)
disorder = group_and_concat(disorder)
```

```python
import os
# 使用するGPUを指定. この環境変数の場所は，pytorchをimportする前に入れる
os.environ["CUDA_VISIBLE_DEVICES"] = "1"
```

```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# トークナイザーとモデルをロード
tokenizer = AutoTokenizer.from_pretrained("Mizuiro-sakura/luke-japanese-large-sentiment-analysis-wrime")
model = AutoModelForSequenceClassification.from_pretrained('Mizuiro-sakura/luke-japanese-large-sentiment-analysis-wrime')

# データフレームの読み込み
df = pd.read_csv('/home/is/sakiho-k/research/data/test_data_user.csv')

# NaNを空文字列に置き換え
df['text'] = df['text'].fillna('')

# 数値を文字列に変換
df['text'] = df['text'].astype(str)

# 感情ラベルの対応
emotions = ['joy、うれしい', 'sadness、悲しい', 'anticipation、期待', 'surprise、驚き', 
            'anger、怒り', 'fear、恐れ', 'disgust、嫌悪', 'trust、信頼']

# 感情スコアを格納するリスト
sentiment_scores = []

# 各テキストに対する感情分析
for text in df['text']:
    # トークン化とモデルへの入力準備
    max_seq_length = 512
    tokens = tokenizer(text,
                       truncation=True,
                       max_length=max_seq_length,
                       padding="max_length",
                       return_tensors="pt")  # Tensor形式で返す

    # モデルに入力
    with torch.no_grad():  # 推論時に勾配計算を無効化
        output = model(input_ids=tokens['input_ids'], attention_mask=tokens['attention_mask'])

    # 最大スコアのインデックスを取得
    max_index = torch.argmax(output.logits, dim=1).item()

    # 感情ラベルを取得
    sentiment_scores.append(emotions[max_index])

# 新しい列を作成して感情スコアを追加
df['sentiment'] = sentiment_scores

# 結果をCSVファイルに保存（オプション）
# df.to_csv('/home/is/sakiho-k/research/data/test_data_user_with_sentiment.csv', index=False)

# 結果の確認
print(df.head())
```

```python
import re
from transformers import pipeline

def split_into_sentences(text):
    # 文末の句点（。）、感嘆符（!）、疑問符（?）やその後に続く空白で分割
    sentences = re.split(r'(?<=[。！？])\s*', text)
    # 文末が空白や改行で終わっている場合に余分な空白を取り除く
    sentences = [s.strip() for s in sentences if s.strip()]
    return sentences

def split_and_merge(text):
    # 文ごとに分割
    sentences = split_into_sentences(text)
    # 各文を改行でさらに分割
    final_sentences = []
    for sentence in sentences:
        lines = sentence.split('\n')
        for line in lines:
            line = line.strip()
            if line:
                final_sentences.append(line)
    return final_sentences

def analyze_sentiment(sentences, sentiment_pipeline):
    results = []
    for sentence in sentences:
        analysis_result = sentiment_pipeline(sentence)
        # 結果からポジティブ、ニュートラル、ネガティブスコアを抽出
        sentiment_scores = {result['label']: result['score'] for result in analysis_result[0]}
        # 最も高いスコアを持つラベルとスコアを選択
        max_label = max(sentiment_scores, key=sentiment_scores.get)
        max_score = sentiment_scores[max_label]
        results.append((max_label, max_score))
    return results

def calculate_average_score(results):
    # 平均スコアを計算するため、各ラベルに対してスコアを集計
    scores = {'positive': 0, 'neutral': 0, 'negative': 0}
    counts = {'positive': 0, 'neutral': 0, 'negative': 0}
    
    for label, score in results:
        if label in scores:
            scores[label] += score
            counts[label] += 1
    
    # 各ラベルの平均スコアを計算
    average_scores = {label: scores[label] / counts[label] if counts[label] > 0 else 0 for label in scores}
    
    # 平均スコアが一つのラベルに偏る可能性があるので、合計の平均を算出
    total_average_score = sum(average_scores.values()) / len(average_scores)
    
    return average_scores, total_average_score

# モデルとトークナイザーの準備
model_name = 'lxyuan/distilbert-base-multilingual-cased-sentiments-student'  # モデル名を指定
sentiment_pipeline = pipeline('sentiment-analysis', model=model_name, return_all_scores=True)

# サンプルテキスト
text = "私はとても辛いです。"

# 文と行の両方で分割
sentences = split_and_merge(text)

# 感情分析
results = analyze_sentiment(sentences, sentiment_pipeline)

# 平均スコアを計算
average_scores, total_average_score = calculate_average_score(results)

# 各分割部分とスコアを表示
for i, (label, score) in enumerate(results, 1):
    print(f"Sentence {i}: {sentences[i-1]}")
    print(f"  Predicted Sentiment: {label}")
    print(f"  Sentiment Score: {score:.4f}")

print(f"\nAverage Scores by Sentiment:")
for label, score in average_scores.items():
    print(f"  {label.capitalize()}: {score:.4f}")

print(f"\nOverall Average Sentiment Score: {total_average_score:.4f}")
```

```python
import torch
from transformers import AutoModelForSequenceClassification, AutoTokenizer
from captum.attr import IntegratedGradients
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from IPython.display import display, HTML

# モデルとトークナイザーの読み込み
model_name = '/home/is/sakiho-k/research/notebooks/model/pretreated/checkpoint-22000'
tokenizer = AutoTokenizer.from_pretrained(model_name, do_lower_case=False)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# サンプル文章
sample_text = "ここに分析したい文章を入力してください。"

# 入力のトークン化
inputs = tokenizer(sample_text, return_tensors='pt')
input_ids = inputs['input_ids']
attention_mask = inputs['attention_mask']

# モデルの出力からlogitsだけを取り出す関数
def custom_forward(input_embeds, attention_mask):
    outputs = model(inputs_embeds=input_embeds, attention_mask=attention_mask)
    logits = outputs.logits
    return logits

# input_idsの埋め込み層を取得
input_embeds = model.get_input_embeddings()(input_ids)
input_embeds.requires_grad_(True)

# IntegratedGradientsのインスタンス化
ig = IntegratedGradients(custom_forward)

# 勾配計算
try:
    attributions = ig.attribute(inputs=input_embeds, additional_forward_args=(attention_mask,), target=0)
    print(attributions)

except Exception as e:
    print(f"Error during attribution calculation: {e}")

# トークンと寄与度を取得
tokens = tokenizer.convert_ids_to_tokens(input_ids[0].tolist())
attributions = attributions.squeeze(0).tolist()  # 1次元リストに変換

# 各寄与度がリストの場合、平均を計算してスカラー値に変換
attributions = [sum(attr) / len(attr) if isinstance(attr, list) else attr for attr in attributions]

# 寄与度に応じて色付けする関数
def colorize_text(tokens, attributions):
    colored_text = []
    
    # 寄与度の最小値と最大値を取得
    min_attribution = min(attributions)
    max_attribution = max(attributions)
    
    # 正規化の設定
    norm = mcolors.Normalize(vmin=min_attribution, vmax=max_attribution)

    for token, attribution in zip(tokens, attributions):
        # 色を決定（白色を避けるため、カラーを選択）
        color_value = norm(attribution)
        color = mcolors.rgb2hex(plt.cm.bwr(color_value)[:3])
        # 白色の場合は黒に変更
        if color_value >= 0.8:  # しきい値を設定
            color = '#000000'  # 黒色
        colored_text.append(f'<span style="color:{color}">{token}</span>')

    return ' '.join(colored_text)

# 色付きテキストの生成
colored_output = colorize_text(tokens, attributions)

# 結果をHTMLとして表示
display(HTML(colored_output))
```

```python
import numpy as np
import torch
import shap
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# モデルとトークナイザーの読み込み
model_name = '/home/is/sakiho-k/research/notebooks/model/all_result/checkpoint-11000'
tokenizer = AutoTokenizer.from_pretrained(model_name, do_lower_case=False)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# モデルを評価モードに設定
model.eval()

# SHAPのためのモデルラッパーを作成
class ModelWrapper:
    def __init__(self, model):
        self.model = model

    def predict(self, texts):
        # SHAP用の入力形式に変換
        inputs = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")
        with torch.no_grad():
            logits = self.model(**inputs).logits
            return torch.nn.functional.softmax(logits, dim=-1).numpy()

# モデルラッパーのインスタンスを作成
wrapped_model = ModelWrapper(model)

# サンプルテキストを準備
sample_texts = [
    "ここに分析したいテキストを入力します。",
    "もう一つのテキストを追加します。",
    "異なるサンプルテキスト。",
]

# SHAP explainerを作成
# sample_texts をそのまま渡す
explainer = shap.KernelExplainer(wrapped_model.predict, np.array(sample_texts))

# SHAP値を計算
shap_values = explainer.shap_values(np.array(sample_texts))

# 最初のサンプルテキストに対するforce plotを表示
shap.initjs()  # Jupyter環境で実行する場合に必要
shap.force_plot(explainer.expected_value[0], shap_values[0], sample_texts[0])
```
