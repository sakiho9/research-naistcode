```python
import pandas as pd
from transformers import BertJapaneseTokenizer, AutoModelForSequenceClassification
import torch
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import numpy as np
import japanize_matplotlib  # 日本語対応を追加

# CSVファイルからtestデータを読み込む
df = pd.read_csv('/home/is/sakiho-k/research/data/test_data_user.csv')

# テキストとラベルを取得
test_texts = df['text'].tolist()
true_labels = df['labels'].tolist()

# モデルとトークナイザーをロード
model = AutoModelForSequenceClassification.from_pretrained('/home/is/sakiho-k/research/notebooks/model/all_result/checkpoint-116')
tokenizer = BertJapaneseTokenizer.from_pretrained('/home/is/sakiho-k/research/notebooks/model/all_result/checkpoint-116', do_lower_case=False)

# トークン化
inputs = tokenizer(test_texts, padding=True, truncation=True, return_tensors="pt", max_length=128)

# モデル予測
model.eval()
with torch.no_grad():
    outputs = model(**inputs)
predictions = torch.argmax(outputs.logits, dim=-1).tolist()

# 混同行列の作成
cm = confusion_matrix(true_labels, predictions)

# ヒートマップの作成と表示
def plot_confusion_matrix(cm, labels, fontsize=12):
    fig, ax = plt.subplots(figsize=(8, 6))
    cax = ax.matshow(cm, cmap='PuBu', alpha=0.75)
    fig.colorbar(cax)

    # 各セルに数値を表示 (フォントサイズ指定)
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(x=j, y=i, s=cm[i, j], va='center', ha='center', fontsize=fontsize)

    # 軸ラベルの設定（フォントサイズ指定）
    ax.set_xticks(np.arange(len(labels)))
    ax.set_yticks(np.arange(len(labels)))
    ax.set_xticklabels(labels, fontsize=fontsize)
    ax.set_yticklabels(labels, fontsize=fontsize)
    plt.xlabel('Predicted', fontsize=fontsize)
    plt.ylabel('True', fontsize=fontsize)
    plt.title('ブログ単位', fontsize=fontsize + 2)
    plt.show()

# ラベルを定義 (例: 0がダイエット、1が摂食障害)
labels = ['Diet', 'Eating Disorder']

# ヒートマップを表示
plot_confusion_matrix(cm, labels, fontsize=20)

# 分類レポートを表示
report = classification_report(true_labels, predictions, target_names=labels)
print("\nClassification Report:")
print(report)
```

# Precision-Recall Curve

```python
import pandas as pd
from transformers import BertJapaneseTokenizer, AutoModelForSequenceClassification
import torch
from sklearn.metrics import precision_recall_curve, auc
import matplotlib.pyplot as plt
import numpy as np

# CSVファイルからtestデータを読み込む
df = pd.read_csv('/home/is/sakiho-k/research/data/test_data_user.csv')  # CSVファイルのパスを指定

# テキストとラベルを取得
test_texts = df['text'].tolist()  # テキストデータ
true_labels = df['labels'].tolist()  # 正解ラベル

# モデルとトークナイザーをロード
model = AutoModelForSequenceClassification.from_pretrained('/home/is/sakiho-k/research/notebooks/model/all_result/checkpoint-116')
tokenizer = BertJapaneseTokenizer.from_pretrained('/home/is/sakiho-k/research/notebooks/model/all_result/checkpoint-116', do_lower_case=False)

# トークン化
inputs = tokenizer(test_texts, padding=True, truncation=True, return_tensors="pt", max_length=128)

# モデル予測
model.eval()  # モデルを評価モードに設定
with torch.no_grad():
    outputs = model(**inputs)
logits = outputs.logits
pred_probs = torch.softmax(logits, dim=-1)[:, 1].tolist()  # ラベル1の予測確率を取得

# Precision-Recall Curve の計算
precision, recall, _ = precision_recall_curve(true_labels, pred_probs, pos_label=1)
roc_auc = auc(recall, precision)  # AUC計算

# Precision-Recall Curve の描画
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, marker='.', color='orange', label=f'PR AUC = {roc_auc:.2f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import torch
from transformers import BertJapaneseTokenizer, AutoModelForSequenceClassification

# モデルとトークナイザーをロード
model_path = "/home/is/sakiho-k/research/notebooks/model/pretreated/checkpoint-22000"
tokenizer = BertJapaneseTokenizer.from_pretrained(model_path, do_lower_case=False)
model = AutoModelForSequenceClassification.from_pretrained(model_path)

# 評価データをCSVから読み込む
data_path = "/home/is/sakiho-k/research/data/pretreated/test_data.csv"
df = pd.read_csv(data_path)

# テキストカラム内の欠損値や非文字列データを除外
df = df.dropna(subset=['text'])  # 欠損値を削除
df = df[df['text'].apply(lambda x: isinstance(x, str))]  # 文字列でない値を除外

# テキストとラベルを再取得
texts = df['text'].tolist()  # テキスト
true_labels = df['labels'].tolist()  # 実際のラベル

# モデルを評価モードに切り替え
model.eval()

# テキストをトークナイズ
inputs = tokenizer(texts, return_tensors='pt', padding=True, truncation=True, max_length=128)

# 予測を実施
with torch.no_grad():
    outputs = model(**inputs)
    logits = outputs.logits
    predictions = torch.argmax(logits, dim=-1).cpu().numpy()  # 予測ラベル

# 予測結果をデータフレームに追加
df['Prediction'] = predictions

# TP, TN, FP, FN の判定列を作成
df['TP'] = ((df['labels'] == 1) & (df['Prediction'] == 1)).astype(int)
df['TN'] = ((df['labels'] == 0) & (df['Prediction'] == 0)).astype(int)
df['FP'] = ((df['labels'] == 0) & (df['Prediction'] == 1)).astype(int)
df['FN'] = ((df['labels'] == 1) & (df['Prediction'] == 0)).astype(int)
```

```python
import pandas as pd

# データの読み込み
df = pd.read_csv('/home/is/sakiho-k/research/data/FP_filtered.csv')

# date列を日付型に変換
df['date'] = pd.to_datetime(df['date'])

# 各ユーザごとに日付順にソート
df = df.sort_values(by=['user', 'date'])

df.head()
```

```python
# データの読み込み
df_2 = pd.read_csv('/home/is/sakiho-k/research/data/pretreated/fp_remaining.csv')

# date列を日付型に変換
df_2['date'] = pd.to_datetime(df_2['date'])

# 各ユーザごとに日付順にソート
df_2 = df_2.sort_values(by=['user', 'date'])

df_2.head()
```

```python
# データの結合
result_df = pd.concat([df, df_2]).sort_values(by=['user', 'date']).reset_index(drop=True)

# ユーザ数の確認
len(result_df['user'].unique())
```

```python
# 日付をdatetime型に変換
result_df['date'] = pd.to_datetime(result_df['date'])

# 各ユーザごとに処理
filtered_df = (
    result_df
    .sort_values(by=['user', 'date'], ascending=[True, False])  # ユーザごとに日付降順でソート
    .groupby('user')  # ユーザごとにグループ化
    .apply(lambda group: group[(group['text'].str.len() >= 300) & (group['text'].str.len() <= 500)].head(1))  # 300～500文字の最初の1行を取得
    .reset_index(drop=True)  # インデックスをリセット
)

print(filtered_df)
```
