```python
# グラフ化
import matplotlib.pyplot as plt
import japanize_matplotlib
import pandas as pd

def plot_bmi_distribution(bmi_data, title):
    """
    BMIデータを棒グラフとして可視化する関数
    
    Parameters:
    - bmi_data: BMIデータのSeries（例: 20代女性、20代男性など）
    - title: グラフのタイトル
    """
    # floatをintに変換
    bmi_data = bmi_data.astype(int)

    # データの分布をカウント
    bmi_counts = bmi_data.value_counts().sort_index()

    # x軸の範囲を指定（10から50まで）
    all_bmi_values = range(10, 51)

    # すべてのBMI値をカバーするように0で補完
    bmi_counts_full = bmi_counts.reindex(all_bmi_values, fill_value=0)

    # グラフ作成
    plt.figure(figsize=(10, 6))
    bmi_counts_full.plot(kind='bar', color='skyblue', edgecolor='black')

    # x軸のラベル設定（5ずつ間引き）
    xticks = range(0, len(bmi_counts_full), 5)  # x軸の目盛り位置
    x_labels = [str(i) for i in range(10, 51, 5)]  # ラベルは10, 15, 20, ..., 50
    x_labels[0] = '10未満'  # 10の位置を「10未満」に変更
    x_labels[-1] = '50未満'  # 50の位置を「50未満」に変更

    # x軸のラベルを更新
    plt.xticks(xticks, x_labels, rotation=0, fontsize=12)

    # グラフの装飾
    plt.title(title, fontsize=16)
    plt.xlabel('BMI', fontsize=14)
    plt.ylabel('Count', fontsize=14)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # グラフを表示
    plt.show()

# グラフ作成関数を各カテゴリに適用
plot_bmi_distribution(bmi_20w, '20代女性')
plot_bmi_distribution(bmi_20m, '20代男性')
plot_bmi_distribution(bmi_30w, '30代女性')
plot_bmi_distribution(bmi_30m, '30代男性')
plot_bmi_distribution(bmi_40w, '40代女性')
plot_bmi_distribution(bmi_40m, '40代男性')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 体型データを数値化する関数
def preprocess_body_data(data_series):
    """
    体型データを数値に変換する関数。
    特定の値を置換し、全角数字を半角に変換して適切に数値化する。
    
    Parameters:
    - data_series: pd.Series (体型データ列)
    
    Returns:
    - pd.Series: 数値化されたデータ
    """
    # 特定の文字列を数値に置換し、全角数字を半角に変換
    data_series = data_series.replace({
        '１より痩せている': '0',
        '８より太っている': '9',
        '１': '1', '２': '2', '３': '3', '４': '4', '５': '5', 
        '６': '6', '７': '7', '８': '8', '９': '9', '０': '0'  # 全角数字を半角数字に変換
    }, regex=True)

    # 数値型に変換
    return data_series.astype(int)

# 散布図を作成する関数
def plot_body_comparison(dataframe, title, current_col=1, ideal_col=2):
    """
    体型データの現在値と理想値を比較する散布図を作成する関数。
    
    Parameters:
    - dataframe: pd.DataFrame (データフレーム)
    - title: str (グラフのタイトル)
    - current_col: int (現在の体型が格納されている列のインデックス)
    - ideal_col: int (理想の体型が格納されている列のインデックス)
    """
    # 現在の体型と理想の体型を抽出し、変換
    current_body = preprocess_body_data(dataframe.iloc[1:, current_col])  # 現在の体型
    ideal_body = preprocess_body_data(dataframe.iloc[1:, ideal_col])  # 理想の体型

    # 散布図を作成
    plt.figure(figsize=(8, 6))
    plt.scatter(current_body, ideal_body, alpha=0.7, color="blue", edgecolor="black")
    plt.plot([0, 9], [0, 9], color='red', linestyle='--', label="X=Y (一致)")

    # ラベルの設定
    plt.title(title, fontsize=16)
    plt.xlabel('現在の体型', fontsize=14)
    plt.ylabel('理想の体型', fontsize=14)
    plt.xticks(range(0, 10), fontsize=12)
    plt.yticks(range(0, 10), fontsize=12)
    plt.grid(alpha=0.5, linestyle='--')
    plt.legend(fontsize=12)
    plt.show()

# 会員ごとに現在の体型と理想の体型の比較ラインプロットを作成する関数
def plot_body_comparison_per_member(dataframe, title, current_col=1, ideal_col=2):
    """
    会員ごとの現在の体型と理想の体型の比較を行うラインプロットを作成する関数。
    
    Parameters:
    - dataframe: pd.DataFrame (データフレーム)
    - title: str (グラフのタイトル)
    - current_col: int (現在の体型が格納されている列のインデックス)
    - ideal_col: int (理想の体型が格納されている列のインデックス)
    """
    # 会員IDを取得
    member_ids = dataframe.iloc[1:, -1]  # 会員ID
    
    # 現在の体型と理想の体型を数値化して格納
    plot_data = pd.DataFrame({
        '会員ID': member_ids,
        '現在': preprocess_body_data(dataframe.iloc[1:, current_col]),  # 現在の体型
        '理想': preprocess_body_data(dataframe.iloc[1:, ideal_col])  # 理想の体型
    })

    # 長い形式に変換
    plot_data_melted = plot_data.melt(id_vars='会員ID', value_vars=['現在', '理想'], 
                                      var_name='カテゴリ', value_name='体型')

    # ラインプロットを作成
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=plot_data_melted, x='カテゴリ', y='体型', hue='会員ID', 
                 marker='o', alpha=0.7, palette='tab10', legend=False)

    # ラベルの設定
    plt.title(title, fontsize=16)
    plt.xlabel('カテゴリ', fontsize=14)
    plt.ylabel('体型', fontsize=14)
    plt.xticks(fontsize=12)
    plt.yticks(range(0, 10), fontsize=12)
    plt.grid(alpha=0.5, linestyle='--')
    plt.show()

# 各データフレームに適用
plot_body_comparison(df_20w, '現在の体型と理想の体型の比較 (20代女性)')
plot_body_comparison(df_20m, '現在の体型と理想の体型の比較 (20代男性)')
plot_body_comparison(df_30w, '現在の体型と理想の体型の比較 (30代女性)')
plot_body_comparison(df_30m, '現在の体型と理想の体型の比較 (30代男性)')
plot_body_comparison(df_40w, '現在の体型と理想の体型の比較 (40代女性)')
plot_body_comparison(df_40m, '現在の体型と理想の体型の比較 (40代男性)')

# 会員ごとの体型比較ラインプロットを作成
plot_body_comparison_per_member(df_20w, '現在の体型と理想の体型の比較 (20代女性 - 会員ごと)')
plot_body_comparison_per_member(df_20m, '現在の体型と理想の体型の比較 (20代男性 - 会員ごと)')
plot_body_comparison_per_member(df_30w, '現在の体型と理想の体型の比較 (30代女性 - 会員ごと)')
plot_body_comparison_per_member(df_30m, '現在の体型と理想の体型の比較 (30代男性 - 会員ごと)')
plot_body_comparison_per_member(df_40w, '現在の体型と理想の体型の比較 (40代女性 - 会員ごと)')
plot_body_comparison_per_member(df_40m, '現在の体型と理想の体型の比較 (40代男性 - 会員ごと)')
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def convert_to_minutes(value):
    """
    '時間'という単位が含まれている場合、時間を分に変換します。
    '分'という単位が含まれている場合、そのまま分として返します。
    それ以外はそのまま返します。
    
    Parameters:
    - value: str or int (SNS使用時間)
    
    Returns:
    - int: 分単位に変換されたSNS使用時間
    """
    if isinstance(value, str):
        # '時間'の場合
        if '時間' in value:
            try:
                hours = int(value.replace('時間', '').strip())  # '時間'を取り除き、整数に変換
                return hours * 60  # 時間を分に変換
            except ValueError:
                return None  # 数値に変換できない場合はNoneを返す
        
        # '分'の場合
        elif '分' in value:
            try:
                return int(value.replace('分', '').strip())  # '分'を取り除き、整数に変換
            except ValueError:
                return None  # 数値に変換できない場合はNoneを返す
        
        # 無効な文字列（例えば'や'など）を除外
        try:
            return int(value)  # 数値として変換可能であればそのまま返す
        except ValueError:
            return None  # 数値に変換できない場合はNoneを返す

    # 数字がそのままの場合はそのまま返す
    try:
        return int(value)
    except ValueError:
        return None  # 数値に変換できない場合はNoneを返す

def remove_outliers(data):
    """
    外れ値を除外する関数 (IQRによる判定)
    
    Parameters:
    - data: list or pandas.Series (数値データ)
    
    Returns:
    - cleaned_data: 外れ値を除外したデータ
    """
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data >= lower_bound) & (data <= upper_bound)]

def plot_bmi_boxplot(data_dict, title):
    """
    SNS使用時間データを箱ひげ図として可視化する関数
    
    Parameters:
    - data_dict: 各カテゴリのデータを格納した辞書（例: {'20代女性': sns_time_20w, ...}）
    - title: グラフのタイトル
    """
    # データをリスト形式に変換し、数値に変換
    data_values = [
        pd.to_numeric(bmi.apply(convert_to_minutes), errors='coerce').dropna() 
        for bmi in data_dict.values()
    ]  # 数値に変換し、欠損値を除外
    
    # 外れ値を除去
    cleaned_data_values = [remove_outliers(bmi) for bmi in data_values]
    
    # グラフ作成
    plt.figure(figsize=(10, 6))
    plt.boxplot(cleaned_data_values, labels=data_dict.keys(), patch_artist=True,
                boxprops=dict(facecolor='lightsteelblue', color='black'),
                medianprops=dict(color='red'))

    # グラフの装飾
    plt.title(title, fontsize=16)
    plt.ylabel('SNS使用時間 (分)', fontsize=14)
    plt.xlabel('年代・性別', fontsize=14)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # グラフを表示
    plt.show()

# 各カテゴリのデータを数値化して格納
data_dict = {
    '20代女性': df_20w_q4[1:],  # 最初の行（設問4）は除外
    '20代男性': df_20m_q4[1:],
    '30代女性': df_30w_q4[1:],
    '30代男性': df_30m_q4[1:],
    '40代女性': df_40w_q4[1:],
    '40代男性': df_40m_q4[1:],
}

# グラフ作成
plot_bmi_boxplot(data_dict, '各年代・性別のSNS使用時間の分布')
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# SNSリスト
sns_list = ['X (旧Twitter)', 'Instagram', 'Facebook', 'Tik Tok', 'BeReal', '該当なし']

# 各カテゴリのSNS使用データ
data_dict = {
    '20代女性': df_20w_q5[1:],  # 最初の行（設問5）は除外
    '20代男性': df_20m_q5[1:],
    '30代女性': df_30w_q5[1:],
    '30代男性': df_30m_q5[1:],
    '40代女性': df_40w_q5[1:],
    '40代男性': df_40m_q5[1:],
}

def count_sns_usage(data, sns_list):
    """
    SNS使用頻度をカウントする関数
    
    Parameters:
    - data: pandas.Series (SNS使用データ)
    - sns_list: 使用されるSNSのリスト
    
    Returns:
    - sns_counts: 各SNSの使用回数
    """
    sns_counts = {sns: 0 for sns in sns_list}  # 各SNSのカウントを初期化
    
    for value in data:
        if isinstance(value, str):
            # '@'で分割して複数のSNSを処理
            sns_entries = value.split('@')
            for sns in sns_entries:
                sns = sns.strip()  # 前後の空白を削除
                for valid_sns in sns_list:
                    if valid_sns in sns:
                        sns_counts[valid_sns] += 1  # 該当するSNSをカウント
    return sns_counts

def plot_sns_usage(data_dict, sns_list):
    """
    SNS使用回数を棒グラフで可視化する関数
    
    Parameters:
    - data_dict: 各カテゴリのSNS使用データを格納した辞書（例: {'20代女性': sns_data, ...}）
    - sns_list: 使用されるSNSのリスト
    """
    sns_usage_by_group = {}
    
    # 各年代・性別のSNS使用頻度をカウント
    for group, data in data_dict.items():
        sns_usage_by_group[group] = count_sns_usage(data, sns_list)
    
    # データフレームに変換
    sns_usage_df = pd.DataFrame(sns_usage_by_group).T  # 行に年代・性別、列にSNS
    sns_usage_df = sns_usage_df[sns_list]  # SNS順に並べ替え

    # グラフ作成
    sns_usage_df.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Set2')

    # グラフの装飾
    plt.title('年代・性別ごとのSNS使用', fontsize=16)
    plt.xlabel('年代・性別', fontsize=14)
    plt.ylabel('使用人数', fontsize=14)
    plt.xticks(rotation=45)
    plt.legend(title="SNS", fontsize=12)
    plt.tight_layout()

    # 凡例の位置を調整（グラフ外に配置）
    plt.legend(title="SNS", fontsize=12, bbox_to_anchor=(1.05, 1), loc='upper left')

    # グラフのレイアウトを調整して、凡例がグラフに重ならないようにする
    plt.tight_layout()

    # グラフを表示
    plt.show()

# SNS使用回数を棒グラフで可視化
plot_sns_usage(data_dict, sns_list)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def calculate_percentage_distribution(data):
    """
    各選択肢の割合を計算する関数
    'いいえと回答' を「ダイエット経験なし」に変換
    
    Parameters:
    - data: pandas.Series (回答データ)
    
    Returns:
    - dict: 各選択肢の割合（キーが選択肢、値が割合）
    """
    # 'いいえと回答' を「ダイエット経験なし」に変換
    data = data.replace('いいえと回答', 'ダイエット経験なし')
    
    total_count = len(data.dropna())  # 有効なデータの総数
    value_counts = data.value_counts()  # 各選択肢のカウント
    return (value_counts / total_count * 100).to_dict() if total_count > 0 else {}

def plot_stacked_bar(data_dict, title, categories):
    """
    ダイエットのモチベーション分布を積み上げ棒グラフで可視化する関数
    
    Parameters:
    - data_dict: 各カテゴリのデータを格納した辞書（例: {'20代女性': data, ...}）
    - title: グラフのタイトル
    - categories: モチベーションのカテゴリリスト
    """
    # 各カテゴリごとの選択肢割合を計算
    percentages = {group: calculate_percentage_distribution(data) for group, data in data_dict.items()}
    
    # 選択肢ごとにデータを整形
    values = {category: [percentages[group].get(category, 0) for group in data_dict.keys()] for category in categories}
    
    # グラフの設定
    x = range(len(data_dict.keys()))  # x軸の位置
    bottom = [0] * len(data_dict)  # 積み上げ棒グラフの下限
    plt.figure(figsize=(10, 6))
    
    # 色味を統一するために colormap 'Set3' を指定
    colormap = plt.get_cmap('Set3')
    colors = [colormap(i / len(categories)) for i in range(len(categories))]  # カラーマップを適用
    
    # 各選択肢を積み上げ棒グラフで表示
    bar_width = 0.6  # 棒の幅を調整
    for idx, (category, vals) in enumerate(values.items()):
        plt.bar(x, vals, bottom=bottom, label=category, width=bar_width, color=colors[idx])
        bottom = [b + v for b, v in zip(bottom, vals)]  # 次の棒の下限を更新
    
    # グラフの装飾
    plt.title(title, fontsize=16)
    plt.xlabel('年代・性別', fontsize=14)
    plt.ylabel('割合 (%)', fontsize=14)
    plt.xticks(x, data_dict.keys(), rotation=45)  # xラベルを斜めに表示
    
    # 凡例の位置を調整（グラフ外に配置）
    plt.legend(title="モチベーション", fontsize=10, loc='upper left', bbox_to_anchor=(1, 1))
    
    # グリッド線
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    # レイアウトを調整して、グラフに収める
    plt.tight_layout()
    
    # グラフを表示
    plt.show()

# ダイエットのモチベーションのカテゴリリスト
motivation_categories = ['理想のモデルの影響', '身の周りの環境', '健康状態の改善（病気予防・治療）', 
                         'スポーツ・パフォーマンス向上', 'その他', 'ダイエット経験なし']

# 各カテゴリのデータ
data_dict = {
    '20代女性': df_20w_q8[1:],  # 最初の行（設問タイトル）は除外
    '20代男性': df_20m_q8[1:],
    '30代女性': df_30w_q8[1:],
    '30代男性': df_30m_q8[1:],
    '40代女性': df_40w_q8[1:],
    '40代男性': df_40m_q8[1:],
}

# グラフ作成
plot_stacked_bar(data_dict, '各年代・性別のダイエットのモチベーション分布', motivation_categories)
```
