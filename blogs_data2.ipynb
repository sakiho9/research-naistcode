## データの読み込み

```python
import pandas as pd

# ダイエットブログ
diet = pd.read_csv('/home/is/sakiho-k/research/notebooks/blog_scraping/diet_blogs.csv')

# 摂食障害ブログ
eating_disorder = pd.read_csv('/home/is/sakiho-k/research/data/eating_disorder_blogs.csv')

# 拒食症ブログ
anorexia = pd.read_csv('/home/is/sakiho-k/research/data/anorexia_blogs.csv')
```

## データの調整

```python
# ダイエットブログの"blog_title"の列を削除
diet = diet.drop('blog_title', axis=1)
diet.head()
```

```python
# 摂食障害ブログの"Unnamed: 0", "disease"の列を削除
eating_disorder = eating_disorder.drop(['Unnamed: 0', 'disease'], axis=1)

# 拒食症ブログの"Unnamed: 0", "disease"の列を削除
anorexia = anorexia.drop(['Unnamed: 0', 'disease'], axis=1)
anorexia.head()
```

```python
# ダイエットブログに"category"列を追加
diet['category'] = 0
diet.head()
```

```python
# 摂食障害ブログの"created_at"の列名を"date"に変更
eating_disorder = eating_disorder.rename(columns={'created_at': 'date'})

# 拒食症ブログの"created_at"の列名を"date"に変更
anorexia = anorexia.rename(columns={'created_at': 'date'})

# 摂食障害ブログに"category"列を追加
eating_disorder['category'] = 1

# 拒食症ブログに"category"列を追加
anorexia['category'] = 1

# 摂食障害ブログと拒食症ブログを結合
disorder = pd.concat([eating_disorder, anorexia], axis=0)

eating_disorder.head()
```

```python
# ダイエットブログと摂食障害ブログ・拒食症ブログを結合
df = pd.concat([diet, disorder], axis=0)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from collections import Counter

# 2. ユーザーごとにテキストを結合
combined_df = pd.concat([diet, disorder])
combined_df['text'] = combined_df['text'].fillna('').astype(str)
grouped_df = combined_df.groupby('user')['text'].apply(' '.join).reset_index()

# 3. テキストの前処理とベクトル化
vectorizer = TfidfVectorizer(max_features=1000, stop_words='english')
X = vectorizer.fit_transform(grouped_df['text'])

# 4. クラスタリング (K-means)
num_clusters = 5  # クラスタ数は任意に設定
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X)
grouped_df['cluster'] = kmeans.labels_

# 5. 結果の可視化
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X.toarray())

plt.figure(figsize=(10, 8))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=grouped_df['cluster'], cmap='viridis')
plt.title('User-Based Cluster Visualization')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(label='Cluster')
plt.show()

# 6. クラスタごとの特徴語と人数の表示
feature_names = vectorizer.get_feature_names_out()

for cluster_num in range(num_clusters):
    # クラスタに属するユーザー
    cluster_users = grouped_df[grouped_df['cluster'] == cluster_num]['user']
    num_users = len(cluster_users)
    
    # クラスタ内のテキストをまとめる
    cluster_texts = grouped_df[grouped_df['cluster'] == cluster_num]['text'].values
    cluster_vector = vectorizer.transform(cluster_texts).sum(axis=0)
    
    # 特徴語を抽出（TF-IDFスコアが高い単語上位10個）
    sorted_items = cluster_vector.argsort()[0, -10:].tolist()[0]
    top_words = [feature_names[i] for i in sorted_items][::-1]  # 上位から表示
    
    print(f"\nCluster {cluster_num}:")
    print(f"Number of users: {num_users}")
    print(f"Top words: {', '.join(top_words)}")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# 2. 必要な列（user, text, category, date）を選択して結合
diet_df = diet[['user', 'text', 'category', 'date']]
ed_df = disorder[['user', 'text', 'category', 'date']]
combined_df = pd.concat([diet, disorder])

# 3. 欠損値の処理とデータ型の統一
combined_df['text'] = combined_df['text'].fillna('').astype(str)
combined_df['date'] = pd.to_datetime(combined_df['date'])  # date列を日付形式に変換

# 4. ユーザーごとに日付順（逆順）でテキストを結合
grouped_df = (combined_df
              .sort_values(by=['user', 'date'], ascending=[True, False])  # ユーザーごとに日付で降順にソート
              .groupby(['user', 'category'])['text']
              .apply(' '.join)  # 各ユーザーの投稿を結合
              .reset_index())

# 5. テキストの前処理とベクトル化
vectorizer = TfidfVectorizer(max_features=1000, stop_words='english')
X = vectorizer.fit_transform(grouped_df['text'])

# 6. クラスタリング (K-means)
num_clusters = 5  # クラスタ数は任意に設定
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X)
grouped_df['cluster'] = kmeans.labels_

# 7. 各クラスタごとのデータフレームを作成
clustered_dfs = {}
for cluster_num in range(num_clusters):
    # 指定された列のみを含める
    cluster_df = grouped_df[grouped_df['cluster'] == cluster_num][['user', 'text', 'category', 'cluster']].copy()
    clustered_dfs[cluster_num] = cluster_df
    print(f"Cluster {cluster_num} DataFrame created with {len(cluster_df)} users")

# 8. クラスタごとのデータフレームを確認（例としてCluster 0を表示）
print(clustered_dfs[0].head())

# 9. 必要に応じてクラスタごとのデータフレームを保存
# 例: clustered_dfs[0].to_csv('cluster_0.csv', index=False)
```

## 最後のブログ5%を兆候ありのブログとする

```python
# ユーザーごとに処理
def update_first_5_percent(df):
    # 「category」が0の行を対象
    target_rows = df[df['category'] == 0]
    
    # 全体の5%の行数を計算
    n_rows_to_update = max(1, int(len(target_rows) * 0.05))
    
    # 最初の5%の行を選択し、「category」を2に更新
    indices_to_update = target_rows.head(n_rows_to_update).index
    df.loc[indices_to_update, 'category'] = 2
    
    return df

# 各「user」ごとに処理
clustered_df_new = clustered_df.groupby('user', group_keys=False).apply(update_first_5_percent)

# 結果を表示
print(clustered_df_new)
```

```python
# clustered_df_newの「category」の値をカウント
category_counts = clustered_df_new['category'].value_counts()
print(category_counts)
```

## 摂食障害ブログの単語頻度

```python
import pandas as pd
from collections import Counter
import MeCab

# MeCabを使って形態素解析し、頻出単語をカウントする
mecab = MeCab.Tagger("-Owakati")  # 単語を分かち書き
all_words = []

for text in disorder['text']:
    words = mecab.parse(text).split()  # 分かち書きした単語をリストに分割
    all_words.extend(words)  # すべての単語を一つのリストに集める

# 単語の出現頻度をカウント
word_counts = Counter(all_words)

# 出現頻度の高い順に上位10個を表示
most_common_words = word_counts.most_common(10)
print(most_common_words)
```
