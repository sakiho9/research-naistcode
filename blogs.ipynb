## 摂食障害ブログ
###### 収集期間 2023年6月〜8月ごろ
###### スクレイピング除外基準 ⏬

###### ・homepageの形式が異なる記事 ( https://ameblo.jp/[user_name]のページが最新記事ではなく，記事一覧となっているもの )

```python
import pandas as pd
# 摂食障害ブログ
eating_disorder = pd.read_csv('/home/is/sakiho-k/research/data/eating_disorder_blogs.csv')
print(eating_disorder.head())

# 拒食症
anorexia = pd.read_csv('/home/is/sakiho-k/research/data/anorexia_blogs.csv')
print(anorexia.head())

# userの人数をグループ化
eating_disorder_user = eating_disorder['user'].groupby(eating_disorder['user']).count().sort_values(ascending=False)
anorexia_user = anorexia['user'].groupby(anorexia['user']).count().sort_values(ascending=False)

print('-'*50)

# userの人数を表示
print(f'摂食障害ブログの人数：{eating_disorder_user.count()}，拒食症ブログの人数：{anorexia_user.count()}')
```
